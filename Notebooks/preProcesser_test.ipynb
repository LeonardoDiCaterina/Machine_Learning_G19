```python
#get the data
train_path = "../Data/train_data.csv"
test_path = "../Data/test_data.csv"
dtype = {'Column29': 'str'}  # Replace 'Column29' with the actual column name that has mixed types

train = pd.read_csv(train_path, index_col='Claim Identifier', dtype=dtype, low_memory=False)
test = pd.read_csv(test_path, index_col='Claim Identifier', dtype=dtype, low_memory=False)
```

```python
# drop all the rows that have missing values in Claim Injury Type
train = train.dropna(subset=['Claim Injury Type'])
y = train['Claim Injury Type']
X = train.drop(columns=['Claim Injury Type'])
# partition the data X, y and y_2bin
X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.2, random_state=0)
```

## start the pipeline
to start the pipeline you need to call the method `pipeline()` giving it the data you want to preprocess

* _note_ : the data should be a pandas dataframe
* _note_ : all the start features should be in the data, if you pass more features than the start features, the extra features will be ignored
<br/> &nbsp;&nbsp;&nbsp;&nbsp; &nbsp;&nbsp;&nbsp;&nbsp; if you pass less features than the start features the preProceser will raise an error
* _note_ : the fit_scale parameter is used to fit the scaler on the data, if you set it to False the scaler will not be fitted on the data
<br/> &nbsp;&nbsp;&nbsp;&nbsp; &nbsp;&nbsp;&nbsp;&nbsp; this is useful when you want to preprocess the test data.
<br/> &nbsp;&nbsp;&nbsp;&nbsp; &nbsp;&nbsp;&nbsp;&nbsp; you don't want to fit the scaler on the test data, you want to fit it on the train data and use it on the test data
* _note_ : when using pipeline on the trainig data you should also set to true the set_end_features parameter
<br/> &nbsp;&nbsp;&nbsp;&nbsp; &nbsp;&nbsp;&nbsp;&nbsp; this is used to set the end features of the data, the end features are the features that will be used in the model

```python
df_train = prep1.pipeline(X_train.copy(), fit_scaler=True, set_end_features=True)
```
